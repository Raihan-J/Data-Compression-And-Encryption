```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()

 
x = np.linspace(0,50,10)

y = np.linspace(0,50,500)

plt.plot(x, np.sin(x),'g', label='sin(x)');
plt.plot(y, np.sin(y),':y', label='sin(x)');
plt.legend();
```
